```python
with open('poems.txt','r') as f:
    data=f.read().split('\n')
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import text
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(oov_token='<nothing>')
tokenizer.fit_on_texts(data)
```

```python
input_sequences=tokenizer.texts_to_sequences(data)
input_sequences
```

```python
inputs=[]
for seq in sequences:
    for i in range(1,len(seq)):
        inputs.append(seq[:i+1])
print(inputs)
```

```python
X=padded_sequences[:,:-1]
y=padded_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=3436)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding,Dense,LSTM
model=Sequential()
model.add(Embedding(3436,output_dim=100,input_shape=(12,)))
model.add(LSTM(150))
model.add(Dense(3436,activation='softmax'))
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
text="moon"
loop=5
for i in range(loop):
    input_=tokenizer.texts_to_sequences([text])[0]
    padded_input=pad_sequences([input_],maxlen=13,padding='pre')
    pos = np.argmax(model.predict(padded_input))
    for word,index in tokenizer.word_index.items():
        if index == pos:
          text = text + " " + word
          print(text)
          time.sleep(2)
```
